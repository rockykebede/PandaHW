```python
# Importing Dependencies
import pandas as pd
import numpy as np
```

```python
# Loading CSV files 
school_data_to_load = "C:/Users/owner/Desktop/GWbootcamp/GWARL201808DATA3/04-Pandas/Homework/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load ="C:/Users/owner/Desktop/GWbootcamp/GWARL201808DATA3/04-Pandas/Homework/PyCitySchools/Resources/students_complete.csv"
```

```python
# Read School data and Student Data inputs and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load,sep=',')
student_data = pd.read_csv(student_data_to_load,sep=',')
school_data
```

```python
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_combined = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_combined.head()
```

```python
# print the head of the dataframes
student_data.head()
```

```python
#Creating a Table that shows the summary of the district's key metrics

summary_keymetrics = pd.DataFrame({ "Total Schools" : school_data_combined['School ID'].nunique(),
"Total Students" : school_data_combined['Student ID'].nunique(),
"Total Budget":school_data['budget'].sum(),"Average Math Score":student_data['math_score'].mean(),"Average Reading Score": student_data ['reading_score'].mean(),"% Passing Math": (student_data['math_score'].apply(lambda x: x >= 70)).mean()*100,
"% Passing Reading": (student_data['reading_score'].apply(lambda x: x >= 70)).mean()*100}, index=[0])
                                  
                  
summary_keymetrics["Overall Passing Rate"] = (summary_keymetrics["% Passing Math"]+ summary_keymetrics["% Passing Reading"] )/2
summary_keymetrics
```

```python
# Creating a Table that shows the school summary of total school budget with total students and Per student budget

school_summary= school_data.groupby(["school_name","type"]).agg({"size":np.sum,"budget":np.sum}).reset_index()
school_summary.columns = ["School Name","School Type","Total Students", "Total School Budget"]
school_summary["Per Student Budget"] = school_summary["Total School Budget"]/school_summary["Total Students"]

school_summary

```

```python
# Creating a table that shows school name, Average Math Score and Average Reading Score using dataframe student_data

school_summary_math_readingscore = student_data.groupby(["school_name"]).agg({'math_score':np.mean,'reading_score':np.mean}).reset_index()
school_summary_math_readingscore
```

```python
# Creating a table percentage of Passing Math and Reading scores 

def passrate(x):return (x['math_score'].apply(lambda x: x >= 70).mean()*100,  
                        x['reading_score'].apply(lambda x: x >= 70).mean()*100)

school_per_math_reading = student_data.groupby('school_name').apply(passrate).reset_index()
school_per_math_reading[['% Passing Math', '% Passing Reading']] = school_per_math_reading[0].apply(pd.Series)
school_per_math_reading = school_per_math_reading[['% Passing Math', '% Passing Reading']]

school_per_math_reading
```

```python
# Final school summary Table showing math, reading score, and percentage of Passing math and reading
Final_school_summary = pd.concat([school_summary,school_summary_math_readingscore,school_per_math_reading],axis=1)
Final_school_summary['Overall Passing Rate']= (Final_school_summary['% Passing Math']+Final_school_summary['% Passing Reading'])/2
Final_school_summary
```

```python
#Table displaing the top five schools in overall passing rate
Final_school_summary.sort_values('Overall Passing Rate',ascending=False)[:5].reset_index()
```

```python
# Table displaing the five worst-performing schools
Final_school_summary.sort_values('Overall Passing Rate',ascending=False)[-5:].reset_index()
```

```python
# Table displaying student math score summary for 9th, 10th, 11th, 12th grades
Student_math_summary = student_data.groupby(['school_name', 'grade']).agg({"math_score":np.mean}).unstack()
Student_math_summary
```

```python
# Table displaying student reading score summary for 9th, 10th, 11th, 12th grades
Student_reading_summary = student_data.groupby(['school_name', 'grade']).agg({"reading_score":np.mean}).unstack()
Student_reading_summary
```

```python
#New dataframe and table showing with spending ranges per student
Spending_students= student_data.merge(school_data,left_on='school_name', right_on='school_name')
Spending_students['Spending Ranges Per Student']=Spending_students['budget']/Spending_students['size']
Spending_students
```

```python
# creating Sample bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
Spending_students['Spending Ranges Per Student'] = pd.cut(Spending_students['Spending Ranges Per Student'], spending_bins ,labels = group_names ,right=False)
```

```python
#creating a table showing Average math and reading scores per student spending ranges
spending_stu_score = Spending_students.groupby(['Spending Ranges Per Student']).agg({'math_score':np.mean,'reading_score':np.mean})
spending_stu_score

                                  
```

```python
# passing percentage and a table showing the passing percentage by Student spending ranges
def passrate(x):return (x['math_score'].apply(lambda x: x >= 70).mean()*100,
            x['reading_score'].apply(lambda x: x >= 70).mean()*100)

spending_stu_score2 = Spending_students.groupby(['Spending Ranges Per Student']).apply(passrate).reset_index()
spending_stu_score2[['% Passing Math', '% Passing Reading']] = spending_stu_score2[0].apply(pd.Series)
spending_stu_score2 = spending_stu_score2.set_index("Spending Ranges Per Student")
spending_stu_score2 = spending_stu_score2[['% Passing Math', '% Passing Reading']]
spending_stu_score2
```

```python
# Merging average math and reading score dataframe with passing percentage dataframe 
#to create a final summary table per studneet spending range
spending_score_summary = pd.concat([spending_stu_score,spending_stu_score2],axis=1)
spending_score_summary ['Overall Passing Rate']= (spending_score_summary['% Passing Math']+spending_score_summary['% Passing Reading'])/2
spending_score_summary
```
